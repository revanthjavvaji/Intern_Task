#Match-Analysis

```python
mat_df.info()
```

```python
pm_counts = mat_df['player_of_match'].value_counts()
pd.DataFrame(mat_df['player_of_match'].value_counts())
```

```python
mat_df['player_of_match'].value_counts().head(10).plot(kind='barh')
plt.ylabel('Player')
plt.xlabel('Number of Player of the Match')
plt.title('Number of Player of the Match Player-Wise(Top)')
plt.show()
```

```python
teams = mat_df.team1.unique()
venues = mat_df.venue.unique()

for team in teams:
    plt.figure()
    for venue in venues:
        team_venue_counts = mat_df[(mat_df['winner'] == team) & (mat_df['venue'] == venue)]
        num_matches_won = team_venue_counts.shape[0]

        plt.bar(venue, num_matches_won)

    plt.xlabel('Venue')
    plt.ylabel('Matches Won')
    plt.title(f"Matches Won by {team} at Each Venue")
    plt.xticks(rotation=90)

    plt.show()
```
